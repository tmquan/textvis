```python
import os 
import glob
import pandas as pd

datadir = "data/physionet.org/mimic-iii-demo/1.4/"

# Read the PATIENTS.csv file
pat_df = pd.read_csv(os.path.join(datadir, 'PATIENTS.csv'), index_col='subject_id')

# Read the ADMISSIONS.csv file
adm_df = pd.read_csv(os.path.join(datadir, 'ADMISSIONS.csv'), index_col='subject_id')
```

```python
# Merge the PATIENTS and ADMISSIONS dataframes on the SUBJECT_ID column
# merged_df = pd.merge(pat_df[['gender', 'dob']],
#                      adm_df[['admittime', 'diagnosis']],
#                      on='subject_id',
#                      how='inner')
merged_df = pd.merge(pat_df,
                     adm_df,
                     on='subject_id',
                     how='inner')
merged_df
```

```python
# Glob all csv files
# filenames = glob.glob(os.path.join(datadir, '*.csv'))
filenames = [
    "ADMISSIONS.csv", 
    "CHARTEVENTS.csv",
    "DIAGNOSES_ICD.csv",
    "ICUSTAYS.csv",
    "LABEVENTS.csv",
    "PATIENTS.csv",
    "PROCEDUREEVENTS_MV.csv",
]
# Step 3: Create an empty DataFrame
merged_df = pd.DataFrame(columns=["subject_id"])

# Step 4: Loop through all files in the folder
for filename in filenames:
    # Step 5: Check if the file is a CSV file
    if filename.endswith(".csv"):
        # Step 5: Read the CSV file into a DataFrame
        df = pd.read_csv(os.path.join(datadir, filename)).dropna()
        # Step 6: Check if the dfFrame has a "subject_id" column
        if "subject_id" in df.columns:
            # Step 7: Merge the dfFrame with the existing merged dfFrame on "subject_id"
            merged_df = pd.merge(merged_df, df, on="subject_id", how="outer")
        else:
            # Step 8: If the DataFrame does not have a "subject_id" column, ignore it
            continue

merged_df
```

```python
merged_df.to_csv(os.path.join(datadir, "MERGED.csv"))
```
